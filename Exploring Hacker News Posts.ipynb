# Exploring Hacker News Posts
### In this project, we will analyze data in order to find out when to post a particular topic on Hack News so that you could attract more comments

## Imports

```python
from csv import reader
from datetime import datetime
```

## Importing datasets

```python
opened_file = open('hacker_news.csv')
read_file = reader(opened_file)
hn = list(read_file)
opened_file.close()
```

```python
hn_header = hn[0]
hn = hn[1:]
```

```python
print(hn_header, '\n')
for i in range(0,5):
    print(hn[i], '\n')
```

##### We will only be concerned with "Ask HN" or "Show HN"
##### We will be creating lists of lists containing just the data for those titles

```python
ask_posts = []
show_posts = []
other_posts = []

for row in hn:
    title = row[1]
    if (title.lower().startswith('ask hn')):
        ask_posts.append(row)
    elif (title.lower().startswith('show hn')):
        show_posts.append(row)
    else:
        other_posts.append(row)
```

```python
print('nr ask posts:', len(ask_posts))
print('nr show posts:', len(show_posts))
print('nr other posts:', len(other_posts))
```

```python
print(hn_header, '\n')
print('First 5 rows of Ask posts')
for i in range(0,5):
    print(ask_posts[i], '\n')
print('First 5 rows of Show posts')
for i in range(0,5):
    print(show_posts[i], '\n')
```

##### Let's determine whether ask_posts or show_posts receive more comments

```python
print("Nr of comments on first ask post:",ask_posts[0][4])
print("Type of this data:", type(ask_posts[0][4]))
```

```python
total_ask_comments = 0
total_show_comments = 0

for row in ask_posts:
    num_comments = int(row[4])
    total_ask_comments += num_comments
avg_ask_comments = total_ask_comments * 1.0 / len(ask_posts)

for row in show_posts:
    num_comments = int(row[4])
    total_show_comments += num_comments
avg_show_comments = total_show_comments * 1.0 / len(show_posts)
```

```python
print('Average nr of comments on Ask_posts:', avg_ask_comments)
print('Average nr of comments on Show_posts:', avg_show_comments)
```

It is showed that the average number of comments on "Ask HN" posts is larger than that of "Show HN" posts. This is because ask posts are more likely to receive answers from other users, addressing the questions from the authors.

##### We shall now determine whether posting an ask post at a certain time will attract more comments. We will perform this analysis via the following steps:
-  Calculate the number of ask posts created in each hour of the day, along with the number of comments received
-  Calculate the average number of comments ask posts receive by hour created

```python
tmp = "December 21, 1984"
tmp = datetime.strptime(tmp, "%B %d, %Y") # %B represents the full name of the month
print(tmp)
```

```python
result_list = []

for row in ask_posts:
    result_list.append([row[6], int(row[4])])
```

```python
result_list[:5]
```

```python
counts_by_hour = {}
comments_by_hour = {}

for row in result_list:
    date_time = datetime.strptime(row[0], '%m/%d/%Y %H:%M')
    hour = date_time.strftime('%H')
    if hour not in counts_by_hour:
        counts_by_hour[hour] = 1
        comments_by_hour[hour] = row[1] #nr of comments
    else:
        counts_by_hour[hour] += 1
        comments_by_hour[hour] += row[1]
```

```python
print(len(counts_by_hour))
print(len(comments_by_hour))
```

```python
avg_by_hour = []

for hour in counts_by_hour:
    avg_by_hour.append([hour, comments_by_hour[hour] * 1.0 / counts_by_hour[hour]])
```

```python
avg_by_hour_desc = sorted(avg_by_hour, key=lambda x : x[1], reverse=True)
template = 'The average nr of comments at {hour} is {cmts:.2f}'
for i in range(0, 5):
    item = avg_by_hour_desc[i]
    print(template.format(hour=item[0], cmts=item[1]), '\n')
```

The posts with most average comments fall within three timestamps: 3PM, 2AM, and 8PM.

We conclude that if you want to attract more attention to your posts and receive more comments, it would be ideal to share your post at either 3PM, 2AM, 8PM, 4PM, or 9PM. You should also consider sharing an "Asking HN" post compared to a "Show HN" discussion.
